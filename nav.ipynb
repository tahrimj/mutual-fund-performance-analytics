```python
import pandas as pd

file_path = "DownloadNAVHistoryReport_sep-dec.txt"

with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()
```

```python
header = lines[0].strip().split(";")

data_lines = [
    line.strip().split(";")
    for line in lines
    if line.strip() and line.strip()[0].isdigit()
]
```

```python
df = pd.DataFrame(data_lines, columns=header)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
print(df.columns)
```

```python
df = df.rename(columns={
    "scheme_code": "scheme_code",
    "scheme_name": "scheme_name",
    "net_asset_value": "nav",
    "date": "date"
})
```

```python
df["nav"] = pd.to_numeric(df["nav"], errors="coerce")
df["date"] = pd.to_datetime(df["date"], format="%d-%b-%Y")
df["scheme_code"] = df["scheme_code"].astype(str)
```

```python
df.to_csv("nav_sep-dec_2025_clean.csv", index=False)
```

```python
import pandas as pd
import glob
import os

# 1. Path to the folder containing your CSVs
path = './clean' 

# 2. Get a list of all CSV filenames
all_files = glob.glob(os.path.join(path, "*.csv"))

# 3. Read each file into a list of DataFrames
df_list = [pd.read_csv(f) for f in all_files]

# 4. Concatenate all DataFrames into one
combined_df = pd.concat(df_list, ignore_index=True)

# 5. Save or view the result
combined_df.to_csv("combined_output.csv", index=False)
print(combined_df.head())
```

```python
df_all = pd.read_csv("combined_output.csv")
```

```python
df_all = df_all.rename(columns={
    "isin_div_payout/isin_growth": "isin_payout_or_growth",
    "isin_div_reinvestment": "isin_reinvestment"
})
```

```python
df_all["scheme_code"] = df_all["scheme_code"].astype(str)
df_all["nav"] = pd.to_numeric(df_all["nav"], errors="coerce")
df_all["date"] = pd.to_datetime(df_all["date"], format='mixed', errors='coerce')
```

```python
df_all["amc"] = (
    df_all["scheme_name"]
    .str.split("Fund", n=1)
    .str[0]
    .str.strip()
)
```

```python
df_all["plan_type"] = df_all["scheme_name"].str.contains(
    "Direct", case=False, na=False
).map({True: "Direct", False: "Regular"})
```

```python
df_all["distribution_type"] = (
    df_all["scheme_name"]
    .str.contains("IDCW|Dividend", case=False, na=False)
    .map({True: "IDCW", False: "Growth"})
)
```

```python
df_all[df_all["scheme_name"].isna()]
```

```python
df_all = df_all.dropna(subset=["scheme_name"])
```

```python
df_final = df_all[
    [
        "scheme_code",
        "scheme_name",
        "amc",
        "plan_type",
        "distribution_type",
        "nav",
        "date"
    ]
]
```

```python
df_final = df_final.sort_values(["scheme_code", "date"])

df_final["daily_return"] = (
    df_final
    .groupby("scheme_code")["nav"]
    .pct_change()
)
```

```python
df_returns = df_final.copy()

df_returns.to_csv(
    "nav_with_daily_returns.csv",
    index=False
)
```

```python
df_returns = df_returns[
    ~df_returns["scheme_name"].str.contains(
        "Unclaimed", case=False, na=False
    )
]
```

```python
selected_funds = [
    "Aditya Birla Sun Life Multi-Cap Fund-Direct Growth",
    "Aditya Birla Sun Life Multi-Cap Fund-Regular Growth",
    "Aditya Birla Sun Life Multi-Cap Fund-Direct IDCW Payout"
]

df_returns = df_returns[
    df_returns["scheme_name"].isin(selected_funds)
]
```
