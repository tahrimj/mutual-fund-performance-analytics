```python
import pandas as pd
benchmark = pd.read_csv("nifty50_historical.csv")
```

```python
benchmark.columns = (
    benchmark.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
benchmark.head()
benchmark["date"].head(10)
```

```python
benchmark["date"] = benchmark["date"].astype(str).str.title()
```

```python
benchmark.head()
benchmark["date"].head(10)
```

```python
benchmark["date"] = pd.to_datetime(
    benchmark["date"],
    format="%d-%b-%Y",
    errors="coerce"
)


benchmark["benchmark_value"] = pd.to_numeric(
    benchmark["close"],
    errors="coerce"
)

```

```python
benchmark = benchmark.sort_values("date")
```

```python
benchmark = benchmark[["date", "benchmark_value"]].dropna()
```

```python
benchmark["benchmark_return"] = benchmark["benchmark_value"].pct_change()
```

```python
benchmark.head(10)
```

```python
benchmark.to_csv("benchmark_processed.csv", index=False)
```
